**Covert '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
into ['Action','Adventure','Fantasy','Science Fiction']**

convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]') is not working as it is string of list we have to convert it into list => use ast.literal_eval(obj)

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['genres']
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
import ast
def convert2(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
movies.iloc[0].crew
```

```python
import ast
def convert3(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if(i['job']=="Director" or i['job']=="Producer" or i['job']=="Writer" or i['job']=="Original Music Composer"):
            L.append(i['name'])
    return list(set(L)) #to remove duplicate names
```

```python
movies['crew']=movies['crew'].apply(convert3)
```

```python
movies['overview']=movies['overview'].apply(lambda x: x if isinstance(x,list) else x.split() if isinstance(x,str)  else []) 
#isinstance() is a built-in Python function used to check the data type of a value. isinstance("hello", str)->True
```

For Genres, Keywords, Cast and Crew remove all spaces between them Sam Worthington[2 entity]->SamWorthington[1 entity] for bag of words

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

**Create new dataframe with id,title,tags remove other cols**

```python
#Convert tags list to string
df.loc[:,'tags']=df['tags'].apply(lambda x: " ".join(x) if isinstance(x,list) else x) 
# .loc->label-based indexer in pandas used to select rows and columns safely.
#df.loc[row,col]->df.loc[:,tags] means label based selection on df on all rows and tags col. Select all rows of the tags column
```

```python
#Convert tags to lower case
df.loc[:,'tags']=df['tags'].apply(lambda x: x.lower())
```

```python
df.head()
```

**Porter Stemmer reduce the words to their root form Ex- "running," "runner," and "ran" can all be reduced to their root form, "run."**

```python
import nltk #Importing Naltural Language Toolkit
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
def root(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)  #Convert list back to string
```

```python
ps.stem('loved')
```

```python
df.loc[:,'tags']=df['tags'].apply(root)
```

```python
df.head()
```

```python
df['tags'][0]
```

**tf-idf vectorizer**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv=TfidfVectorizer(max_features=5000,stop_words='english')
vector=tv.fit_transform(df['tags']).toarray()
```

```python
vector.shape            
#No of movies 4806 transform it acc to 5000 features(here words)  each row represents 1 movie
#vector is the numerical representation of your movie dataset created by TF-IDF from the tags column. 
#Each movie’s text is converted into numbers so that a machine learning algorithm can understand it.
#The vectorizer selected the top 5000 most important words from all movie tags
#Each column represents one unique word (feature)
#Each movie represented as 5000 dim vector
```

```python
vector[0]
```

```python
tv.get_feature_names_out().tolist()
```

**Cosine Similarity**

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
similarity.shape
```

```python
#Index of movies
df[df['title']=='Batman Begins'].index.item()
```

```python
#Sort movies so that most recommended movies are shown upwards as sorted in desc order but index of the recommended movie is lost 
sorted(similarity[1].tolist(),reverse=True)
```

```python
#To keep both index and sort 
#sorted(enumerate((similarity[0]))) -> gives enuumerate object
sorted(enumerate(similarity[1].tolist()),reverse=True) #Sorting done on basis of index
```

```python
#Sorting on basis of 2nd parameter i.e. similarity keeping index also
sorted(enumerate(similarity[1].tolist()),reverse=True,key=lambda x:x[1])
```

```python
#From the above list extract 5 most recommended movies i.e. index 12,199,17,216,3570 here
sorted(enumerate(similarity[1].tolist()),reverse=True,key=lambda x:x[1])[1:6]
```

```python
#Define function to find 5 most recommended movies
def recommend(movie):
    #if movie in df['title'].values:
    movie_index=df[df['title']==movie].index.item()
    dist=similarity[movie_index]
    recommend_movies=sorted(enumerate(dist.tolist()),reverse=True,key=lambda x:x[1])[1:6]
    for i in recommend_movies:
        print(df.iloc[i[0]].title)     #i is tuple contains (movie_index,similarity); iloc=integer location used to select rows and columns by index
```

```python
recommend('Batman Begins')
```

```python
recommend('Avatar')
```

```python
recommend('Iron Man')
```
